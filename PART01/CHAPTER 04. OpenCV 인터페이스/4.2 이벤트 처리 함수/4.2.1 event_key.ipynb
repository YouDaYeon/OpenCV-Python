### 키 이벤트 사용

```python
import numpy as np
import cv2

## switch case문을 사전(dictionary)으로 구현
switch_case = {
    ord('a'): "a키 입력",                  # ord() 함수 문자 -> 아스키 코드 변환
    ord('b'): "b키 입력",
    0x41 : "A키 입력",
    int('0x42', 16) : "B키 입력",          # 0x42(16진수) -> 10진수 변환
    2424832 : "왼쪽 화살표 입력",          # 0x250000
    2490368 : "윗쪽 화살표 입력",          # 0x260000
    2555904 : "오른쪽 화살표 입력",        # 0x270000
    2621440 : "아래쪽 화살표 입력",        # 0x280000
}

image = np.ones((200, 300), np.float)      # 원소값 1인 행렬 생성
cv2.namedWindow("Keyboard Event")
cv2.imshow("keyboard Event", image)

while True:                                # 무한 반복
    key = cv2.waitKeyEx(100)               # 100ms 동안 키 이벤트 대기
    if key == 27: break                    # ESC 누르면 종료

    try:
        result = switch_case[key]
        print(result)
    except KeyError:
        result = -1

cv2.destroyAllWindows()                    # 열린 모든 윈도우 제거
```

> Colab용 수정된 코드

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 1. 키 이벤트를 대신할 입력 딕셔너리 (switch-case 역할)
switch_case = {
    'a': "a키 입력",
    'b': "b키 입력",
    'A': "A키 입력",
    'B': "B키 입력",
    'left': "왼쪽 화살표 입력",
    'up': "윗쪽 화살표 입력",
    'right': "오른쪽 화살표 입력",
    'down': "아래쪽 화살표 입력",
    'esc': "ESC키 입력 - 종료"
}

# 2. OpenCV에서 사용할 이미지 생성 (Colab에서는 matplotlib 사용)
image = np.ones((200, 300), dtype=np.float32)  # float 대신 float32 사용
image *= 255  # 밝기 조절

# 3. Colab에서는 `cv2.imshow()`를 사용할 수 없으므로 `matplotlib`로 이미지 출력
plt.imshow(image, cmap='gray')
plt.title("Keyboard Event")
plt.axis('off')
plt.show()

# 4. 키 입력을 받는 방식 변경 (cv2.waitKeyEx() 대신 input() 사용)
while True:
    key = input("Press a key (a, b, A, B, left, up, right, down, esc to exit): ").strip().lower()

    if key == 'esc':  # 종료 조건
        print("프로그램 종료")
        break

    # 5. switch-case 방식으로 입력된 키를 처리
    result = switch_case.get(key, "지원되지 않는 키 입력")
    print(result)
```
