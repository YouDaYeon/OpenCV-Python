### 4.2.2 마우스 이벤트 제어

```python
import numpy as np
import cv2

def onMouse(event, x, y, flags, param):          # 콜백 함수 - 이벤트 내용 출력
    if event == cv2.EVENT_LBUTTONDOWN:
        print("Left down")
    elif event == cv2.EVENT_RBUTTONDOWN:
        print("Right Down")
    elif event == cv2.EVENT_RBUTTONUP:
        print("Right up")
    elif event == cv2.EVENT_LBUTTONDBLCLK:
        print("Left Double Click")

image = np.full((200, 300), 255, np.uint8)

title1, title2 = 'Mouse Event1','Mouse Event1'
cv2.imshow(title1, image)
cv2.imshow(title2, image)

cv2.setMouseCallback(title1, onMouse)             # 마우스 콜백 함수
cv2.waitKey(0)
cv2.destroyAllWindows()
```

> Colab용 수정된 코드

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 1. 마우스 이벤트를 시뮬레이션하는 함수
def onMouseSimulated(event_type):
    events = {
        'left_down': "Left Button Down",
        'right_down': "Right Button Down",
        'right_up': "Right Button Up",
        'left_dbl_click': "Left Double Click"
    }
    print(events.get(event_type, "Unknown Event"))

# 2. 흰색 배경의 이미지 생성
image = np.full((200, 300), 255, dtype=np.uint8)

# 3. Colab에서는 `cv2.imshow()`를 사용할 수 없으므로 `matplotlib`로 이미지 출력
plt.imshow(image, cmap='gray')
plt.title("Mouse Event Simulation in Colab")
plt.axis('off')
plt.show()

# 4. 사용자 입력을 받아 마우스 이벤트 시뮬레이션
while True:
    event_type = input("Enter a mouse event (left_down, right_down, right_up, left_dbl_click, exit to quit): ").strip().lower()

    if event_type == 'exit':  # 종료 조건
        print("프로그램 종료")
        break

    # 5. 시뮬레이션된 마우스 이벤트 실행
    onMouseSimulated(event_type)
```
