### 4.2.3 트랙바 이벤트 제어

```python
import numpy as np
import cv2

def onChange(value): # 트랙바 콜백 함수
    global image, title # 전역 변수

    add_value = value -int(image[0][0]) # 트랙바 값과 영상 화소값의 차
    image = image + add_value # 행렬과 스칼라 덧셈
    cv2.imshow(title, image)

image = np.zeros((300, 500), np.uint8) # 영상 생성

title = "trackbar event"
cv2.imshow(title, image)

cv2.createTrackbar("Brightness", title, image[0][0], 255, onChange) # 트랙바 콜백
cv2.waitKey(0)
cv2.destroyAllWindows()
```

> Colab용 수정된 코드

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# 1. 초기 흑백 이미지 생성
image = np.zeros((300, 500), dtype=np.uint8)

# 2. Matplotlib으로 이미지 출력하는 함수
def show_image(img):
    plt.figure(figsize=(6, 4))
    plt.imshow(img, cmap='gray')
    plt.title("trackbar event")
    plt.axis('off')
    plt.show()

# 3. 트랙바(슬라이더) 값이 변경될 때 실행될 함수
def onChange(value):
    global image  # 전역 변수 사용
    add_value = value - int(image[0, 0])  # 트랙바 값과 기존 화소값 차이 계산
    image = np.clip(image + add_value, 0, 255)  # 값 제한 (0~255)
    show_image(image)  # 이미지 다시 표시

# 4. 슬라이더(트랙바) 생성
slider = widgets.IntSlider(value=0, min=0, max=255, step=1, description="Brightness")
widgets.interactive(onChange, value=slider)

# 5. UI 표시
display(slider)

# 6. 초기 이미지 출력
show_image(image)
```
