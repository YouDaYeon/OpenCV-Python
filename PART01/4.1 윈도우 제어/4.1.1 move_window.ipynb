ColabÏóêÏÑúÎèÑ OpenCVÎ•º ÏÇ¨Ïö©Ìï¥ÏÑú Ïù¥ÎØ∏ÏßÄÎ•º Ï≤òÎ¶¨ÌïòÍ≥† ÌëúÏãúÌïòÎäî Î∞©Î≤ï

üëâ cv2.imshow() ÎåÄÏã† matplotlibÏùÑ ÏÇ¨Ïö©

```python
# ÎÑòÌååÏù¥ ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏûÑÌè¨Ìä∏
# OpenCV ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏûÑÌè¨Ìä∏

# ÌñâÎ†¨ ÏÉùÏÑ±
# Î∞ùÏùÄ ÌöåÏÉâ(200) Î∞îÌÉï ÏòÅÏÉÅ ÏÉùÏÑ±

# ÏúàÎèÑÏö∞ Ïù¥Î¶Ñ
# ÏúàÎèÑÏö∞ ÏÉùÏÑ± Î∞è ÌÅ¨Í∏∞ Ï°∞Ï†ï ÏòµÏÖò

# ÏúàÎèÑÏö∞ Ïù¥Îèô - ÏúÑÏπò ÏßÄÏ†ï

# ÌñâÎ†¨ ÏõêÏÜåÎ•º ÏòÅÏÉÅÏúºÎ°ú ÌëúÏãú

# ÌÇ§ Ïù¥Î≤§Ìä∏(key event) ÎåÄÍ∏∞
# Ïó¥Î¶∞ Î™®Îì† ÏúàÎèÑÏö∞ ÌååÍ¥¥
```

```python
import numpy as np
import cv2

# ÌñâÎ†¨ ÏÉùÏÑ±
# Î∞ùÏùÄ ÌöåÏÉâ(200) Î∞îÌÉï ÏòÅÏÉÅ ÏÉùÏÑ±
image = np.zeros((200, 400), np.unit8)
image[:] = 200

# ÏúàÎèÑÏö∞ Ïù¥Î¶Ñ
# ÏúàÎèÑÏö∞ ÏÉùÏÑ± Î∞è ÌÅ¨Í∏∞ Ï°∞Ï†ï ÏòµÏÖò
title1, title2 = 'Position1', 'Position2'
cv2.namedWindow(title1, cv2.WINDOW_AUTOSIZE)
cv2.namedWindow(title2)

# ÏúàÎèÑÏö∞ Ïù¥Îèô - ÏúÑÏπò ÏßÄÏ†ï
cv2.moveWindow(title1, 150, 150) # Î™®ÎãàÌÑ∞Ïùò 150, 150 ÏúÑÏπòÏóê ÏúàÎèÑÏö∞ ÏúÑÏπò
cv2.moveWindwo(title2, 400, 50) # Î™®ÎãàÌÑ∞Ïùò 400, 50 ÏúÑÏπòÏóê ÏúàÎèÑÏö∞ ÏúÑÏπò

# ÌñâÎ†¨ ÏõêÏÜåÎ•º ÏòÅÏÉÅÏúºÎ°ú ÌëúÏãú
cv2.imshow(title1, image)
cv2.imshow(title2, image)

# ÌÇ§ Ïù¥Î≤§Ìä∏(key event) ÎåÄÍ∏∞
cv2.waitKey(0)
# Ïó¥Î¶∞ Î™®Îì† ÏúàÎèÑÏö∞ ÌååÍ¥¥
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Î∞ùÏùÄ ÌöåÏÉâ(200) Î∞îÌÉïÏùò ÏòÅÏÉÅ ÏÉùÏÑ±
image = np.zeros((200, 400), dtype=np.uint8)  # uint8Î°ú Î≥ÄÍ≤Ω
image[:] = 200
```

**üëâ uint8ÏùÄ Ïù¥ÎØ∏ÏßÄÎ•º ÌëúÌòÑÌïòÎäî ÌëúÏ§Ä Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ(0~255)**

np.zeros()Îäî Í∏∞Î≥∏Ï†ÅÏúºÎ°ú float64 ÌÉÄÏûÖÏúºÎ°ú Î∞∞Ïó¥ÏùÑ ÏÉùÏÑ±ÌïúÎã§.

ÌïòÏßÄÎßå, OpenCVÏóêÏÑúÎäî Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞Í∞Ä Î∞òÎìúÏãú unit8 ÌÉÄÏûÖÏù¥Ïñ¥Ïïº ÌïúÎã§.
