```python
import numpy as np
import cv2

image = np.zeros((200, 300), np.unit8)  # ndarray 행렬 생성
image.fill(255)                               # 모든 원소에 255(흰색) 지정

title1, title2 = 'AUTOSIZE', 'NORMAL'         # 윈도우 이름 변수
cv2.namedWindow(title1, cv2.WINDOW_AUTOSIZE)  # 윈도우 생성 - 크기 변경 불가
cv2.namedWindow(title2, cv2.WINDOW_NORMAL)    # 크기 변경 가능

cv2.imshow(title1, image)                     # 행렬 원소를 영상으로 표시
cv2.imshow(title2, image)
cv2.resizeWindow(title1, 400, 300)            # 윈도우 크기 변경
cv2.resizeWindow(title2, 400, 300)
cv2.waitKey(0)                                # 키 이벤트(key event) 대기
cv2.destroyAllWindows()                       # 열린 모든 윈도우 제거
```

- 'AUTOSIZE' 윈도우는 행렬(이미지)의 크기 변경없이 윈도우 크기만 바꿈.
- 'NORMAL' 윈도우는 변경된 윈도우와 동일하게 행렬(이미지)의 크기도 바뀐다.
 - image 행렬의 실제 크기가 아니라 보여지는 형태만 바꾸는 것.

> Colab용 수정된 코드

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 1. 흰색(255) 이미지를 생성 (dtype=np.uint8 설정)
image = np.zeros((200, 300), dtype=np.uint8)
image.fill(255)

# 2. Colab에서는 OpenCV의 GUI 함수 대신 matplotlib을 사용해야 함
# AUTOSIZE 모드와 NORMAL 모드의 차이를 나타내기 위해 두 가지 크기로 표현
image_resized = cv2.resize(image, (400, 300))  # 크기 변경

# 3. 이미지 출력
fig, axes = plt.subplots(1, 2, figsize=(8, 4))  # 2개의 서브플롯 생성

axes[0].imshow(image, cmap='gray')  # AUTOSIZE (원본 크기)
axes[0].set_title("AUTOSIZE (Original Size)")
axes[0].axis('off')

axes[1].imshow(image_resized, cmap='gray')  # NORMAL (크기 변경)
axes[1].set_title("NORMAL (Resized to 400x300)")
axes[1].axis('off')

plt.show()
```
